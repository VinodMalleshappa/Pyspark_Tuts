```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Tutorial3").getOrCreate()
spark
```

```python
#Read the data
df = spark.read.csv("top_4000_movies_data.csv", header=True, inferSchema=True)
```

```python
df.show()
```

```python
df=df.withColumnRenamed("Production Budget","Production_Budget")
df.show()
```

```python
#df.withColumn("Age Of the Movie",(Release Date-current_timestamp()).show()

from pyspark.sql.functions import current_timestamp
df = df.withColumn("As on Date",current_timestamp())
df.show()
```

```python
df = df.withColumnRenamed("As on Date","curr_timestamp")
df.show()
```

## Filter Operation

```python
# Production Budget is less than or qual to 260000000 Rs
df.filter("Production_Budget<= 260000000").show()
```

```python
# As we  can see the Production_Budget<= 260000000 Rs and we can fetch perticuler column
df.filter("Production_Budget<= 260000000").select("Movie Title","Release Date","Production_Budget").show()
```

```python
#Another way for filter the columns is 
df.filter(df["Worldwide Gross"]>= 1654367425).show()
```

```python
# Multiple filter condition using "And" Operator
df.filter(
         (df["Production_Budget"]>= 260000000)
         &
         (df["Domestic Gross"]>=543638043)
         ).show()
```

```python
#Not Operator  '~'  
df.filter(~(df["Production_Budget"]<=260000000)).show()
# Using Not Operator, shows that Production_Budget is not less than or equal to 2600000Rs
```
